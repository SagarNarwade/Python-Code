```python
n = int(input())
str1 = set(map(int,input().split()))

operations = int(input())
for i in range(0,operations):
    op = input().split()
    op = op[0]
    str2 = set(map(int,input().split()))
    
    def inter(x,y):
        x.intersection_update(y)
        
    def update(x,y):
        x.update(y)
        
    def symmetric(x,y):
        x.symmetric_difference_update(y)
        
    def difference(x,y):
        x.difference_update(y)
        
    dispatcher = { 'intersection_update' : inter, 'update' : update, 'symmetric_difference_update':symmetric,'difference_update':difference}

    def call_func(x, y, func):
        try:
            return dispatcher[func](x, y)
        except:
            return "Invalid function"

    call_func(str1,str2,op)
print(sum(str1))
    
    
    
    
    
```
